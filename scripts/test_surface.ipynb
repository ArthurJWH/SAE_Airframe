## Testing surface generation

```julia
using MyPackage.Geometry
using MyPackage.VLM
using MyPackage.MyPlots
```

```julia
esa40 = Airfoil("../assets/airfoils/ESA40/ESA40.dat")
wing = Surface(airfoils=[esa40, esa40], b=6.0, chord=y->2-y, sweep=y->30, sw_center=0.0, dihedral=y->30*y^7)

plain = Airfoil("../assets/airfoils/Plain/Plain.dat")
h_stab = Surface(airfoils=[plain, plain], b=2.0, pos=(5.0,0.0,0.0), chord=y->(1-y^2)^0.5, sweep=y->0, sw_center=0.5, dihedral=y->0)
v_stab = Surface(airfoils=[plain, plain], b=1.0, pos=(5.0,0.0,0.0), chord=y->1-0.5*y, sweep=y->0, sw_center=1.0, dihedral=y->0, vertical=true, mirror_xz=false)
```

```julia
plane = Plane([wing, h_stab, v_stab])
```

```julia
mesh = VLMMesh(plane, 32, 32)
```

```julia
p = plot_mesh(mesh)
```
